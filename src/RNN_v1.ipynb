## Model testing

### Splitting the data

```python
from sklearn.model_selection import train_test_split

feature_col_names=['P2', 'P5', 'P6', 'P1_1', 'P1_2', 'P3_1', 'P3_2', 'P4_1', 'P4_2',
       'P4_3', 'P4_4', 'P4_5', 'P4_6', 'P4_7', 'P4_8', 'P4_9', 'P4_10',
       'P4_11', 'P4_12', 'P4_13', 'P4_14', 'P4_15', 'P4_16', 'P4_17', 'P4_18',
       'P4_19', 'P4_20', 'P4_21', 'P4_22', 'P4_23', 'P4_24', 'P7_1', 'P7_2',
       'P7_3', 'P7_4', 'P7_5', 'P7_6', 'P7_7', 'P7_8', 'P7_9', 'P7_10',
       'P7_11']
predicted_class_name=['C']

features_columns = df[feature_col_names]
target_column = df[predicted_class_name]

X = features_columns.values
y = target_column.values
## we split the data 70% training set and 30% test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print('train:', X_train.shape)
print('test:', X_test.shape)
```

```python
print("{0:0.2f}% in training set".format((len(X_train)/len(df.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(df.index)) * 100))
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Model Building - ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

def model_builder(optimizer):
    model = Sequential()
    model.add(Dense(
        units=5, kernel_initializer='uniform',
        activation='relu', input_dim=len(features_columns.columns))) ##number of columns minux the target column= 42
    model.add(Dense(
        units=5, kernel_initializer='uniform',
        activation='relu'))
    model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics= ['accuracy'])
    return model

model = model_builder('adam')
```

#### Model Fitting

```python
model.fit(X_train, y_train, batch_size=5, epochs=100)
```

#### Classifier

```python
classifier = KerasClassifier(build_fn=model_builder)
params = {
    'batch_size': [1, 20],
    'epochs': [10],
    'optimizer': ['adam', 'rmsprop']
}
gs = GridSearchCV(estimator=classifier, param_grid=params,
                 scoring='accuracy', cv=5)
gs = gs.fit(X_train, y_train)
best_params = gs.best_params_
best_score = gs.best_score_
```
