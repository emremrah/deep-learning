# Hastalık Veri Kümesi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hastalik_veriseti/data.csv", sep=';')
print(data.loc[:, 'P1'].value_counts())
print(data.loc[:, 'P3'].value_counts())
print(data.loc[:, 'P4'].value_counts())
print(data.loc[:, 'P7'].value_counts())
```

```python
X_1 = data.iloc[:, :len(data.columns) - 1]
y = data.iloc[:, len(data.columns) - 1]
```

```python
X = pd.get_dummies(X_1, columns=['P1', 'P3', 'P4', 'P7'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print('train:', X_train.shape)
print('test:', X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

def model_builder(optimizer):
    model = Sequential()
    model.add(Dense(
        units=5, kernel_initializer='uniform',
        activation='relu', input_dim=len(X.columns)))
    model.add(Dense(
        units=5, kernel_initializer='uniform',
        activation='relu'))
    model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics= ['accuracy'])
    return model

model = model_builder('adam')
model.fit(X_train, y_train, batch_size=5, epochs=100)
```

```python
predictions = model.predict(X_test)
predictions = (predictions > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predictions)
print(cm)
```

```python
classifier = KerasClassifier(build_fn=model_builder)
params = {
    'batch_size': [1, 20],
    'epochs': [100],
    'optimizer': ['adam', 'rmsprop']
}
gs = GridSearchCV(estimator=classifier, param_grid=params,
                 scoring='accuracy', cv=5)
gs = gs.fit(X_train, y_train)
best_params = gs.best_params_
best_score = gs.best_score_
```

```python
print(best_params)
print(best_score)
```
